```python
import pandas as pd
from pathlib import Path
import os

# 1. Trova la root del progetto in modo assoluto e blindato
current_dir = Path(os.getcwd())
if current_dir.name == 'scripts':
    project_root = current_dir.parent
else:
    project_root = current_dir

cartella_processed = project_root / 'data' / 'processed'

print(f"Sto leggendo dalla cartella corretta: {cartella_processed}")

# 2. Leggi TUTTI i database Parquet
df_cosmos = pd.read_parquet(cartella_processed / 'cosmos_soil_moisture.parquet')
df_topo = pd.read_parquet(cartella_processed / 'topography.parquet')
df_gps = pd.read_parquet(cartella_processed / 'electrode_gps.parquet')
df_seq_elec = pd.read_parquet(cartella_processed / 'sequence_electrodes.parquet')
df_seq_meas = pd.read_parquet(cartella_processed / 'sequence_measurements.parquet')

# 3. Esporta in CSV per la visualizzazione umana
df_cosmos.to_csv(cartella_processed / 'VISUALIZZA_cosmos.csv')
df_topo.to_csv(cartella_processed / 'VISUALIZZA_topografia.csv')
df_gps.to_csv(cartella_processed / 'VISUALIZZA_electrode_gps.csv')
df_seq_elec.to_csv(cartella_processed / 'VISUALIZZA_sequence_electrodes.csv')
df_seq_meas.to_csv(cartella_processed / 'VISUALIZZA_sequence_measurements.csv')

print("\nSUCCESSO. Vai nella cartella data/processed/ nell'albero dei file.")
print("Ho generato 5 file 'VISUALIZZA_...'.")
```
